```python
#feature scaling is a method to scale numeric feature in same scale/range(ex->-1 to 1,0 to 1)
#ie;suppose some heights are in cm and some are in ft,,,100cm and 8ft  ,,but m/c cant understand unit so it will tell 100cm>8ft
#m/c understand only Numbers
#so feature scaling is important
#this is last step involved in data preprocessing and before ML model training,,,also called data normalization
```

```python
#we apply feature scaling on independent variable
#we fit feature scaling with train data and transform on train and test data..(VVIMP)
```

```python
#why we dont use sc.fit() or mmc.fit() on X_test data ?
#->bcz,if we use this sc.fit() or mmc.fit() on X_test data then mean and std deviation (for std scaler) and min and max (for min max scaler)
#is calculated w.r.t X-test data,,thus X_train data has diff scaling and X_test data has diff scaling,,and therefore 
#ml algo not give that much accuracy
#so,,,we train and transform the X_train data and calculated mean and deviation for sc and min and max for mmc by fit() function is stored in sc or mmc object respectively
#thus we transform calculated mean and deviation or min and max to X_train and X_test
#therefore we can use sc.fit_transform() or mmc.fit_transform() for X_train and we use calculated mean and deviation or min and max 
#to feature scale the X_test data ,,where calculated mean and deviation or min and max stored in sc or mmc object respectively
#where sc is StandardScaler object and mmc is MinMaxScaler object
#and same terminology applied to other feturescaling methods,,ie;fit() is applied to X0train and transform is aplied to both X_train and X_test
```

```python
#standardisation-->(StandardScaler)
    #rescale feture such as mean(M)=0,standard deviation(@--)=1 of each feature
    #result of standardisation is Z ,called as z-score normalization
   # z=(x(i)-(mean))/(deviation)  .i=0 to n  ie;no of rows in feature.calculating z value for each value(x(i)) fro each feature
```

```python
'''if data follow normal distribution(guassian distribution) then use this,,(not neccessory that use this only)
if original dta distribution is normal(bell curve/symetric curve),then the standardised distribution will be normal
if original dta distribution is skewed(asymmetric),then the standardised distribution also be skewed

'''
```

```python
#Normalization-->(minmaxscaling)
    #rescale feture in fixed range b/n" 0 to 1 "
    #X(norm)=(X-Xmin)/(Xmax-Xmin)...for ach value of each feature
    #if data doesnt follow normal distribution use this(not neccessory to use this only in this case also)
```

```python
#there is no any thumb rule to use Standardisation or normalization for special ML algo
#But mostly Standardisation use for clustering analysis,principal component analysis(pca)
#NORMALIZATION prefer for image processing bcz,pixel intensity b/n 0 to 255 ,NEURAL algo require data in scale 0-1,k-nearest neighbors
```

# Feature scaling

# standardisation and normalization

```python
#applicable in numeric var and should be independent var
#feature scaling should not be applied on dependent var
```

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df2=df[['survived','pclass','age','parch']]
df2.head()
```

```python
df3=df2.fillna(df2.mean())
df3
```

```python
#survived is dependent var(o/p var) so drop it
#x means i/p and y means o/p
X=df3.drop('survived',axis=1)#bcz inplace=False so,,,temporarily changed in temporary variable
y=df3['survived']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=51)#test_size=0.2 means 
#,20% from current data is used for test and 80% data is used for train
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

```python
#now our data is ready for feature scaling
sc=StandardScaler()
sc.fit(X_train)#here StandardScaler will learn and calculate the mean and standard deviation from X_train
```

```python
sc.mean_
#to see  calculated mean of X_train  (3 columns present)
```

```python
sc.scale_
#to see  calculated standard deviation of X_train  (3 columns present)
```

```python
X_train.describe()#calculate and show the info about X_train,,as u see below
```

```python
#now transform the value learned/calculated by standardscaler  into X_train and X_test
X_train_sc=sc.transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
X_train_sc
```

```python
X_train_sc=pd.DataFrame(X_train_sc,columns=['pclass','age','parch'])
X_test_sc=pd.DataFrame(X_test_sc,columns=['pclass','age','parch'])
```

```python
X_train_sc.head()
```

```python
# succefully scaled X_train and X-test
```

```python
X_train_sc.describe()
#
```

```python
X_train_sc.describe().round(2)#as u see mean and std deviation is 0 and 1 respectively
#round(2),,bcz values are in float and in the  powers of negative ,so round down them up to 2 decimal point,,thus we get desired result
#thus we get numeric value with decimal exactly 2(after point only 2 value)
#so after fit and transform using StandardScaler mean and std deviation of each feature is 0 and 1 respectively
```

# minmaxscaler

```python
mmc=MinMaxScaler()
mmc.fit(X_train)#on X_train minmaxscaler is fitted and learned the minimum and maximum value for each feature of X_train
```

```python
X_train_mmc=mmc.transform(X_train)
X_test_mmc=mmc.transform(X_test)
```
