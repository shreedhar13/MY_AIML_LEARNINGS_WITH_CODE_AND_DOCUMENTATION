```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

```python
#create own points
x=6*np.random.rand(200,1)-3#create random no bn 0 to 1 and return array(numpy array) with dimension 100 row and 1 column
y=0.8*x**2+0.9*x+2+np.random.rand(200,1)#quadratic eq ,reture numpy array
plt.scatter(x,y,color='b')
plt.xlabel('x data')#see carefully,,on x axis negative values are there represents,puting integer both negative,positive value in eq so that parabola is formed
plt.ylabel('y data')
```

```python
#linear regression model with no poynomial term
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
lr=LinearRegression()
lr.fit(X_train,y_train)#training the model
y_pred=lr.predict(X_test)
r2_score(y_test,y_pred)#as u see nearer to 0 ,it is worst model,,
```

```python
plt.scatter(X_train,y_train)#actual training value
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')#model crated after training,,ie;after training ,eq is formed(simple linear eq) 
#in that eq putting the respective features value of training data(X_train) to get this line ,,called regression line/prediction line/ linear regression model
#and same eq is used while executing lr.predict(X_test),,at that time X_test data is putted in eq and comparing with trained line
#it will predict the value
#internally it uses trained line to predict


#below one is our LINEAR REGRESSION MODEL,,,,created while training
```

```python
plt.plot(X_train,lr.predict(X_train),color='r')
plt.scatter(X_train,y_train)
#model with not good fit line
```

```python
plt.plot(X_test,y_pred,color="r")#predicted value
plt.scatter(X_test,y_test)#actual value
```

```python
#error is too high so model is not accurate so use polynomial regression
```

# Polynomial linear regression

```python
#add degree2 terms
poly=PolynomialFeatures(degree=2)
X_train_trans=poly.fit_transform(X_train)#polynomial terms are calculated(by fit) and  added to X_train (by transform) and resultant dataset assigned to X_train_trans
X_test_trans=poly.transform(X_test)#same calculated terms are added to X_test also and resultant dataset assigned to _test_trans

#now X_train_trans is used to train the our model(polynomial reg model)
```

```python
print(X_train[0])
print(X_train_trans[0])#if u dont want 1. ie;x(0) column then include_bias=False in PolynomialFeatures(include_bias=False)
```

```python
lr=LinearRegression()
lr.fit(X_train_trans,y_train)#model is trained and learned the attributes and properties of dataset and thus 
#create the eq of polynomial lin reg(ie;curve) and create a curve,,,(best fit curve
#now ,when u test the data using predict() function that time above curve is reffered to predict o/p for new value(X_test)
```

```python
y_pred=lr.predict(X_test_trans)#dont test X_test othrwise u get error-->ValueError: X has 1 features, but LinearRegression 
#is expecting 3 features as input.
```

```python
r2_score(y_test,y_pred)
#now accuracy is increased,,around 98% accuracy.in linear model 23% accuracy is there is above
```

```python
#as u see r2 score is nearer to 1,,that means model is generalized.actual value and predicted value(curve ) 
#is very closer to each other ,,hence accuracy is high
#ie;error is less
```

```python
print(lr.coef_)#bcz 3 feature are there,,,,a0 X(0)+a1 X(1)+a2 X(2)    X(1)== X^1
print(lr.intercept_)
```

```python
#VVVVVVVVVVVVVIIIMMMPPP
# 0.8x^2 + 0.9x +2
# a2     + a1   +intercept   as u see value is nearly matching,,,eq to predict o/p,,,and we generated o/p using this eq only,,so how model works
#ie;Model means ,,which generate eq of line/curve ..etc by seeing o/p and i/p variable,,ie;estimate the relationship thus regression
```

```python
#but why calculated coef and intercept are not perfect as we use y= 0.8x^2 + 0.9x +2
#bcz we added noise to it  np.random.rand(200,1)
```

```python
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)#bcz we are testing ,,so just using values calculated at the time of training X_train_trans
y_new=lr.predict(X_new_poly)
#bcz X_train and X_test is shuffled while train_test_split ,,so that if u going to plot
#plt.plot(X_train,lr.predict(X_train_trans),"r-",linewidth=1,label='prediction') that time u get zig zagness(bcz data is not in order so...),,you can see by applying below
#thats why we are creating a "dummy values" in a same scale as that of data,,and making it AS another test dataset
#and thus we are not done any train_test_split on these dummy data set so,,they are in sorted manner,so they can give us the prediction line/model for visualizing,,dont 
#it will not effect our actual data ,"IT IS TEST DATA ONLY BROOO>>>>"

#suppose u have real world data ,,and if u dont want to create dummy values like this for visualizing graph(bcz u have to scale according to ur actual data)
#so,,,intially convert it to data frame ,,so that it has indexing and thus after train_test_split ,data is shuffled,,we can sort it according to INDEX..thus u dont get zig zag while visualizing
```

```python
plt.plot(X_new,y_new,"r-",linewidth=1,label='prediction')#r-  means join points with line(-)
plt.plot(X_train,y_train,"y.",linewidth=2,label='training points')#y. and b.  dot means scatter plot
plt.plot(X_test,y_test,"b.",linewidth=2,label='testing points')
plt.xlabel('x')
plt.ylabel("y")
plt.legend()
plt.show()
#genearalized model
```
