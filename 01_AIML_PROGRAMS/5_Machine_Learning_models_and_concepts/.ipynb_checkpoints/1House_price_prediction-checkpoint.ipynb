# Business problem-->predict the price of Banglore house

# using linear regression-supervised ml algo (linear regression model)

multiple linear regression

```python
import pandas as pd
```

```python
df=pd.read_csv(r'https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB')
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
df.head()
```

```python
#as u see dataset is preprocessed already,,so u can directlt train now
```

```python
df.shape
```

```python
X=df.drop('price',axis=1)#means drop price named column(ie;7120 rows of that column also deleted)#remaining are independent var,,1 means row wise operation,all rows are of price col is droped
y=df['price']#dependent var
print('shape of x=',X.shape)
print('shape of y=',y.shape)
```

```python
#now (X,y) is our data set,,so split it into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)#note untill random_state=1,,how many times u execute this statement that many times it will generate same train and test dataset as it is there while executing 1st time
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

```python
#20% of data is randomly splitted as test and 80% of data is randomly splitted as train
#randomly means,rows are selected randomly for train and test,,not sequentially
```

```python
#as u see above feature/column has numeric data of diff unit,,ie;bath and  in float b/n 0 to 10 represent no of bathrooms in houce,same for balcony
#price in Lakhs,area in sqft and..etc....and m/c just understand  units so,,"feature scaling" is required
```

# FEATURE SCALING

```python
#u can use any method but we are using StandardScaler
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)#u can directly use sc.fit_transform(X_train),,,not for X_test  (we just transform the calulated data from X_train by fit() To X_test,,so that train and test data become compatible and have same values like mean and std deviation )
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)

#ie;sc object contains the mean and std deviation of each feature when we use fit()
#and transform() will calculate the z score for each value(x(i)) of each feature(f(i)) -->z(i)=(x(i)-mean)/(std deviation)
# """ f1 | f2 | f3 | f4
# _______|____|____|___________
#     x1 | x1 | x1 | x1
#     x2 | x2 | x2 | x2
#     x3 | x3 | x3 | x3
#     .  |  . | .  |.
# """
```

```python
X_train
```

```python
features=df.columns.drop('price')
X_train=pd.DataFrame(X_train,columns=features)
X_test=pd.DataFrame(X_test,columns=features)
X_test.head()
```

```python
#as u see all features are scaled,,now its time to train the model using linear model
```

# linear regression-ML model training Phase
#from LinearRegression class u can achieve the both simple and multiple linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()#object creation
lr.fit(X_train,y_train)#this will train the ml model,ie;calculate intercept(alph 0),regression coefficient(alpha 1,alpha 2,...etc)
#X_train have 107 features so 107 regression coefficient are calculated and stored in lr object,
#ie;machine will learn how y_train is related to X_train's all features thus calculate regression coefficient and intercept and stored in lr object
```

```python
lr.score(X_train,y_train)*100#accuracy of training data 
#ie;created prediction line/regression line using Training data (X_train) is compared with actual value of
#traing data(y_train),,ie;how model(pred line) is fitted with training so that testing data is predicted more accurately
#ie;how accurate our model is build (prediction line)
#ie;if predicted value and actual value on same pt then it is 100% accurate,,if all predicted values lies/intersect
#the respective actual value then it is OVERFITTING case ,that means we are done some mistake in X_train data
#thus while testing the model for unseen data(X_test) ,that time accuracy is decreased more..see handwritten BOOK(overfitting,underfitting,general fitting)
#so if accuracy of traing data is 100% then use Ridge and lasso model to make it less,,so that it can perform prediction more accurately when it see unseen data)X_test or new data
#othrwise prediction are very bad/worst
```

```python
#m/c learned the data succefully..
#now check what coefficient are learned by our ml model
```

```python
lr.coef_
#(alpha1 to alpha 107)  ie; regression coef for each feature is calculated
```

```python
lr.intercept_
#y=c+m1 x1+m2 x2+m3 x3+m4 x4+......  or y=a0 + a1 x1+a2 x2+.....a=alpha,a0=intercept,a1 to a107=regression coef
```

# ML model testing Phase

```python
#as u see our model is 90% accurate while predicting price with X_test/unseen data 
#we can deploy the model,,but if u want to increase the accuracy then use "hyper parameter tuning","again cleaning data",doing some changes ,or use diff ml algo model
#we can increase accuracy
```

# Implementing Ridge and Lasso Regression

```python
#to increase accuracy,and avoid overfitting and feature selection
#Ridge and lasso is a type of linear regression only
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
rd=Ridge()
rd.fit(X_train,y_train)#traing the model using Ridge regression
rd.score(X_train,y_train)
```

```python
rd.score(X_test,y_test)
```

```python
#as u see accuracy is same as linear regresion so increase alpha value,,testing accuracy is increased very lttle
```

```python
ls=Lasso()
ls.fit(X_train,y_train)
ls.score(X_train,y_train)
#little decreased compare to Ridge
```

```python
ls.score(X_test,y_test)
#little increased compare to Ridge
```

```python
#now apply alpha value,,,alpha=any positive integer 0 to n
```

```python
rd2=Ridge(alpha=2)
rd2.fit(X_train,y_train)#traing the model using Ridge regression
print(rd2.score(X_train,y_train))
print(rd2.score(X_test,y_test))
```

```python
ls2=Lasso(alpha=2)
ls2.fit(X_train,y_train)
print(ls2.score(X_train,y_train))
print(ls2.score(X_test,y_test))
```

```python
ls2=Lasso(alpha=3)
ls2.fit(X_train,y_train)
print(ls2.score(X_train,y_train))
print(ls2.score(X_test,y_test))
```

```python
#as u see after some point if u increase alpha,,then accuracy of testing data is decreasing,
#bcz,,Ridge and lasso change the steepness of regression line,, so there is increase in dist bn actual data pts and regression line
#so accuration is decreased,so carefully take alpha value
#and see handwritten notes for more info
```

# Model Evaluation

```python
#to check error(cost function or MSE or RMSE) in model so that we can select best model
#thats why we use rmse to find correct model by checking its error value
```

```python
y_pred=lr.predict(X_test)
y_pred#predicted value
```

```python
y_test
```

```python
#as u see there is error..ie;dist bn actual value and predicted value is high
#so take average error use MSE or RSME
```

# MSE AND RMSE

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
print('mean square error=',mse)
print('Root mean square error=',rmse)
```

```python
#ie;average diff bn actual value and predicted value is 28 laks rupees
#ex-->house manager saying house price is 100 lakhs but our model saying 100-28=72 lakhs rupees or 100+28=>128 lakh,,,this is the error or cost function
#ex-->house price is 34 lakhs but model saying 34-28=6 lakhs or 34+28=>62 lakh

```

# we can tell that this model is UNDERFITTED

```python
#so this musch error we dont want,,so use the PREPROCESSING on dataset again or applay other model whose error is low as 
#possible(not 0 or closer to 0,,then model is overfitt)
#and that model should not be underfitted also like above linear regression model
#we want Generalized model-->ie;error is minimum as possible(not 0 or closer to 0,,then model is overfitt)
```

```python
#and we deploy our model in real world only if its error is less,,so that it can predict very closer value
```

# Polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg=PolynomialFeatures(degree=2)
X_train_poly=poly_reg.fit_transform(X_train)
X_test_poly=poly_reg.transform(X_test)
```

```python
X_train_poly.shape,X_test_poly.shape
#internally this musch features are created
```

```python
X_train_poly
```

```python
lr=LinearRegression()
lr.fit(X_train_poly,y_train)#takes time bcz (5696, 5886) and 5696 of y_train
```

```python
lr.score(X_test_poly,y_test)#it is internally uses r2 so,in that case best score is that which is nearer to 1 and bad score is nearer to 0
#w.R.T  test data
#as u score accuracy is nearly 100% on test datat(unseen data)
```

```python
lr.score(X_train_poly,y_train)
```

```python
y_test
```

```python
y_pred=lr.predict(X_test_poly)
y_pred
```

```python
#as u see predicted value is exactly same as actual value of test data
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

mse,rmse
```

```python
#as u see rmse is not near to 0 soo,,model is not overfit and rmse is not too high so that it is not underfit,,,instead it is Generalized model
```
