```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB")#bcz data is already cleaned,and onehot encoding is done to convert categorical data into integer,,start working on it
df.head()
```

```python
X=df.drop('price',axis=1)#independent features
y=df['price']#dependent features
X.shape,y.shape#7120 rows and 1 column,,dont think that 7120 are index and 0 column...note->index are not a part of data frame ie;it is not a specifuc feature.it is internal representation of rows and column
```

```python
#bcz indexing is there so after train_test_split() u can sort data using index so that at the time of plotting u cant get zig zag line
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)#note untill random_state=1,,how many times u execute this statement that many times it will generate same train and test dataset as it is there while executing 1st time
print('shape of X_train=',X_train.shape)
print('shape of y_train=',y_train.shape)
print('shape of X_test=',X_test.shape)
print('shape of y_test=',y_test.shape)
```

```python
#if u just want to sort index,,,ie;rows are not sorted according to index  ,,they stay as it is ,,only index is sort ,then use df.reset_index(drop=True)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(X_train)#u can directly use sc.fit_transform(X_train),,,not for X_test  (we just transform the calulated data from X_train by fit() To X_test,,so that train and test data become compatible and have same values like mean and std deviation )
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.svm import SVR  #support veactor regression ,s for classifier use SVC
```

```python
#c -->regularization parameter ie;ridge,lasso  ,epsilon->distance bn marginal plane...etc parameters are there in  SVR()
```

```python
svr_rbf=SVR()#by default kernel=rbf
svr_rbf.fit(X_train,y_train)# best fit hyperplane and marginal plane calculated
svr_rbf.score(X_test,y_test)# above calculated hyperplane(prediction line) used to predict values
#NOTE_>SVM dont have transform function
```

```python
svr_rbf.predict(X_test)
```

```python
y_test
```

```python
#40% acuracy
```

```python
svr_linear=SVR(kernel='linear')
svr_linear.fit(X_train,y_train)
svr_linear.score(X_test,y_test)
#accuracy is increased
```

```python
y_pred=svr_linear.predict(X_test)
y_pred
```

```python
y_test
```

```python
svr_poly=SVR(kernel='poly',degree=2)
svr_poly.fit(X_train,y_train)
svr_poly.score(X_test,y_test)
#accuracy decreased drastically
```

```python
svr_poly.predict(X_test)
```
